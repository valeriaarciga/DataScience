4. Generate a boxplot of sepal_width grouped by species. For each boxplot calculate lower and upper whiskers, Q1, Q3, and median.

```python
#creacion de figuras
plt.figure(figsize=(8, 5))
#grafica
sns.boxplot(x='species', y='sepal.width', data=iris, color='hotpink')
#personalizacion y muestra
plt.title("Boxplot de Sepal Width por especie")
plt.show()
```

```python
#seleccion de la columna
col = "sepal.width"

#desarrollo de tabla
stats = iris.groupby("species")[col].describe()[["25%", "50%", "75%"]]
stats.rename(columns={"25%": "Q1", "50%": "Median", "75%": "Q3"}, inplace=True)

#compute IQR
stats["IQR"] = stats["Q3"] - stats["Q1"]

#compute whiskers
stats["Lower Whisker"] = iris.groupby("species")[col].apply(lambda x: x[x >= (stats.loc[x.name, "Q1"] - 1.5 * stats.loc[x.name, "IQR"])].min())
stats["Upper Whisker"] = iris.groupby("species")[col].apply(lambda x: x[x <= (stats.loc[x.name, "Q3"] + 1.5 * stats.loc[x.name, "IQR"])].max())

#display the result
print(stats)
```

5. In the previous boxplot, overlay a swarmplot to look at individual data points

```python
#creacion de figura
plt.figure(figsize=(8, 5))
#creacion de la grafica
sns.boxplot(x='species', y='sepal.width', data=iris, color='hotpink')
#personalizacion
sns.swarmplot(x='species', y='sepal.width', data=iris, color='darkmagenta')
plt.title("Boxplot + Swarmplot de Sepal Width")
plt.show()
```
